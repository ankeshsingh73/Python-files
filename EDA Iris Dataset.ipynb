```python
"""CDF: Cummulative distribution function"""
# 1 ...... 11 ....... 21 ........ 31 ....... 41 ....... 51

# |---10---|----10----|----10-----|----10----|----10----|

# so we have found out the bin edges now

# to find thte counts we calcuate how many number of points fall into each bin

# therefore the count of a bin = number of elements of a such that left_bin_egde<=ai 5 [1,2,3,5,10]

# ii. number of elements belongs to the 2nd bin 11<=x<21 => 5 [11,12,13,14,15]

# iii. number of elements belongs to the 3rd bin 21<=x<31 => 7 [21,23,25,26,27,29,30]

# iii. number of elements belongs to the 3rd bin 21<=x<31 => 7 [21,23,25,26,27,29,30]

# iv. number of elements belongs to the 4th bin 31<=x<41 => 2 [31,35]

# v. number of elements belongs to the 5th bin 41<=x<=51 => 1 [51]

# note: from the documentation: https://docs.scipy.org/doc/numpy/reference/generated/numpy.histogram.html

# All but the last (righthand-most) bin is half-open i.e [1,2,3,4], the bins are [1,2), [2,3), [3,4]

# [1,10) = 1,2,3,4,5,6,7,8,9 means includig 1 and but not 10. its half open bracket

a= [1, 2, 3, 5, 10,11, 12, 13, 14, 15, 21,23, 25, 26, 27, 29, 30,31, 35, 51]

counts,bins=np.histogram(a)#by default bin size is 10
print (type(counts))
print (type(bins))
print ("counts in each bin starting from bin 0 to last bin= {} and bin edges ={}".format(counts,bins))

print ('_______________________________________________________________________________')
counts,bins=np.histogram(a,bins=5)
print (type(counts))
print (type(bins))
print ("counts in each bin starting from bin 0 to last bin= {} and bin edges ={}".format(counts,bins))

print ('_______________________________________________________________________________')

counts,bins=np.histogram(a,bins=7)
print (type(counts))
print (type(bins))
print ("counts in each bin starting from bin 0 to last bin= {} and bin edges ={}".format(counts,bins))


"""Difference between pdf and cdf:-

CDF tells you the the probability of an independent variable for its own given value
where as PDF tells you how many values an independent variable has in a give window

area under the curve of a pdf given you the cdf at for that point. 


"""
```

```python
 """creating a cdf and pdf for variety- virginicia
 
 to get cdf you need to have an array that has values for petal length(say) for flower virginica
 """
iris_virginica=iris.loc[iris['variety']=='Virginica']
iris_virginica.head(4)
```

```python
"""lets plot pdf of irs-virginica"""
sns.FacetGrid(data=iris_virginica,hue='variety',height=4).map(sns.distplot,'petal.length').add_legend()
```

```python
counts,bin_edges=np.histogram(iris_virginica['petal.length'],bins=10,density=True)
print (("Counts in each of the bins {}".format(counts)))
print ("edges of each bin".format(bin_edges))
pdf=counts/sum (counts)
print (pdf)
```

```python
cdf=np.cumsum(pdf) #cummulative sum a function in numpy
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.xlabel('virginica_petal.length')
plt.ylabel('Probability from 0 to 1')
plt.show()
```

```python
"""PLotting PDF and CDF taking petal.length as independent variable for all feilds of variety"""

iris_setosa=iris.loc[iris['variety']=='Setosa']
iris_versicolor=iris.loc[iris['variety']=='Versicolor']
iris_virginicia=iris.loc[iris['variety']=='Virginica']
#plot for setosa
print ('setosa')
counts,bin_edges=np.histogram(iris_setosa['petal.length'],bins=10,density=True)
print ("counts{}".format(counts))
print ('edges {}'.format(bin_edges))
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)


#plot for Versicolor

print ('Versicolor')
counts,bin_edges=np.histogram(iris_versicolor['petal.length'],bins=10,density=True)
print ("counts{}".format(counts))
print ('edges {}'.format(bin_edges))
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

#plot for Virginica

print ('Virginica')
counts,bin_edges=np.histogram(iris_virginica['petal.length'],bins=10,density=True)
print ("counts{}".format(counts))
print ('edges {}'.format(bin_edges))
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)

plt.xlabel('Petal length for all variety')
plt.ylabel('Probability')


plt.show()


```

```python
"""Median'
Unlike the mean which gets affected by a simple outlier, Median in other place gives exact same value like mean, 
to calculate median- Sort the numbers in ascending or descending order and then choose the middle one(in case of odd count of number)

In case of even count of numbers, sort them and then take the mean of two numbers present in the middle. 

A median will be unaffected by the outliers till the count of outlier is minimum, 

"""
print (np.median(iris_setosa['petal.length'])) #median of the petal length of setosa. 
print (np.mean(iris_setosa['petal.length'])) #mean of the petal length of setosa 
print (np.median(np.append(iris_setosa['petal.length'],500))) #even the outlier does not change the value,




```
