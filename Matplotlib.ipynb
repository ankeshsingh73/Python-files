```python
"""Subplot and fucntion introduction in matplotlib"""
"""Subplot creates different plots in a figure based on paramenter"""


def plotfunc(l1):
    return (np.sin(l1))
def plotcos(l1):
    return (np.cos(l1))
l1=np.arange(0,np.pi*2,0.01) 
plt.figure(1)
plt.subplot(211)  #211-2 rows , single column, figure 1
plt.plot(l1,plotfunc(l1))

# plt.figure(2)
plt.subplot(212)
plt.plot(l1,plotcos(l1), 'rs')
```

```python
"""For better understanding of subplot"""

plt.figure(1)
plt.subplot(211)
plt.plot([2,16,4],'b--')

plt.subplot (212)
plt.plot([5,25,5])


plt.figure(2)
plt.subplot(211)
plt.plot([10,1000,500],'r--')

plt.subplot (212)
plt.plot([15,225,5])

plt.figure(1) #if you want to go back to figure 1
plt.subplot(2,1,1) #plot 1 of figure1
plt.title('Factors of 2') #any kind of changes can be done later by refereing the figure and the plot
```
