```python
"""Confidence interval training using bootstrap technique"""
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
#load dataset
x=np.array([180,162,158,172,168,150,171,188,172,123])

#configure bootstrap
n_size=1000
x_size=int(len(x))

#run bootstrap
medians=list()
for i in range(n_size):
    s=resample(x,n_samples=x_size) #create a sample from x with size of sample (m) less than 
    m=np.median(s)
    medians.append(m)
#plotting part
plt.hist(medians)


```

```python
alpha=0.95 #let say you want 95% of cofidence interval
p1=((1-alpha)/2)*100 #2.5%
lower= np.percentile(medians,p1) #lower median range

p2=(alpha+((1-alpha)/2))*100#95+2.5=97.5%
upper=np.percentile (medians,p2) #upper median range

print ("{} % confidence interval of median is between {} and  {} ".format(alpha*100,lower,upper))

```

```python
alpha=0.90 #let say you want 95% of cofidence interval
p1=((1-alpha)/2)*100 #2.5%
lower= np.percentile(medians,p1) #lower median range

p2=(alpha+((1-alpha)/2))*100#95+2.5=97.5%
upper=np.percentile (medians,p2) #upper median range

print ("{} % confidence interval of median is between {} and  {} ".format(alpha*100,lower,upper))

```

```python
""" if we are resampling from our sample, how is it that we are learning something about the population
rather than only about the sample?
Since the sample is the only information we have about the population, \
we take our sample itself as a model of population.
In this case, resampling is not done to provide an estimate of the population distribution. Rather,
resampling is done to provide an estimate of the sampling distribution of the sample statistic in question."
"""
```

```python
"""K-S Test"""
#generating a 1000 samples for a random variable X having gaussian distribution 
x=stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
print(type(x))
sns.kdeplot(x,bw=0.5)
#mean =0 and variance =1
plt.show()
```

```python
stats.kstest(rvs=x,cdf='norm')#pvalue is high, we assume x is a gaussian distributed
```
